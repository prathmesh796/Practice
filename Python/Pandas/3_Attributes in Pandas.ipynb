A view is copy of the existing data. We can create a view of the data in pandas as follows:

```python
d2 = d1
```

```python
d2
```

Any updation done in d2 we also affect the original dataframe.

```python
d2[0][1] = 3
```

```python
d1
```

```python
d2
```

If we want to create a copy of the existing dataframe then we can use .copy() method provided by DataFrames in Pandas.

```python
d1_copy = d1.copy()
```

```python
d1_copy
```

```python
d1_copy[0][1] = 10
```

```python
d1_copy
```

```python
d1
```

whenever we make any changes or updates in the dataframe, it is not gauranteed that the device will return a copy or a view of that dataframe. So to do any updation we use loc() function

```python
d1.loc[1, 'B'] = 56
```
